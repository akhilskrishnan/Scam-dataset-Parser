```python
from tkinter import *
import pandas as pd

finli=[]
noli=[]
temp=[]
save=False
availableFin=0
availableNo=0

top = Tk()
top.geometry("800x700")
def open_text():
    global finli
    global temp
    global availableFin
    global btn_text3
    availableFin+=1
    btn_text3.set("YES: "+str(availableFin)+" NO: "+str(availableNo)+" available: "+str(len(temp)))
    finli.append(temp.pop())
    text_file=tex=temp[len(temp)-1]
    mytext.delete(1.0,END)
    mytext.insert(END,text_file)
def open_textno():
    global noli
    global temp
    global availableNo
    global btn_text3
    availableNo+=1
    btn_text3.set("YES: "+str(availableFin)+" NO: "+str(availableNo)+" available: "+str(len(temp)))
    noli.append(temp.pop())
    text_file=tex=temp[len(temp)-1]
    mytext.delete(1.0,END)
    mytext.insert(END,text_file)
def save_file():
    global finli
    global noli
    global temp
    global save
    global btn_text2
    global availableFin
    global availableNo
    global btn_text3
    old_hostile=pd.read_csv("hostile.csv",encoding="utf8")#encoding='cp1252')
    old_normal=pd.read_csv("normal.csv",encoding="utf8")#encoding='cp1252')
    if(save):
        old_hostile=old_hostile['0'].to_list()
        new_hostile=pd.DataFrame(old_hostile+finli)
        new_hostile.to_csv("hostile.csv",index=False)
        old_normal=old_normal['0'].to_list()
        new_normal=pd.DataFrame(old_normal+noli)
        new_normal.to_csv("normal.csv",index=False)
        temp.reverse()
        xs=pd.DataFrame(temp)
        xs.to_csv("unfinfilteredData.csv",index=False)
        finli=[]
        noli=[]
        temp=[]
        save=False
        availableFin=0
        availableNo=0
        btn_text3.set("")
        mytext.delete(1.0,END)
        btn_text2.set("LOAD")
    else:
        availableFin=len(old_hostile)+1
        availableNo=len(old_normal)+1
        temp=pd.read_csv("unfinfilteredData.csv",encoding="utf8")#encoding='cp1252')
        temp=temp['0'].to_list()
        btn_text3.set("YES: "+str(availableFin)+" NO: "+str(availableNo)+" available: "+str(len(temp)))
        temp.reverse()
        tex=temp[len(temp)-1]
        mytext.insert(END,tex)
        mytext.pack(pady=10)
        save=True
        btn_text2.set("SAVE")
        
    
open_button = Button(top, text = "YES" ,command=open_text)
open_button.pack(pady =20)
open_buttonn = Button(top, text = "NO" ,command=open_textno)
open_buttonn.pack(pady =20)
btn_text2 = StringVar()
open_buttons = Button(top, textvariable=btn_text2 ,command=save_file)
btn_text2.set("LOAD")
open_buttons.pack(pady =20)
btn_text3 = StringVar()
label = Label(top, textvariable = btn_text3, bg = "green", bd = 100, fg = "white", font = "Castellar")
label.pack(pady=20,side = LEFT) 
open_buttons.pack(pady =20)



mytext = Text(top, width=700,height=700)


  
mainloop()  
 
```
